LECTURE 45: Splitting,Applying and Combining

```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame
```

```python
dframe_wines = pd.read_csv('winequality-red.csv',sep=';')
dframe_wines
```

```python
def ranker(df):
    df['alcohol_rank'] = np.arange(len(df))+1
    return df
```

```python
dframe_wines.sort_values(by='alcohol',ascending=False,inplace=True)
dframe_wines
```

```python
dframe_wines = dframe_wines.groupby('quality').apply(ranker)
```

```python
pd.set_option('display.max_rows',None)
dframe_wines
```

By this the rows are ranked based on their quality
i.e for each quality there is separate rank(eg: quality-5 separate rank,quality6- separate rank etc.)

LECTURE 46: Cross Tabulation

```python
import pandas as pd
from io import StringIO
```

```python
data = """\
Sr.no Animal Intelligence
1 Dog Smart
2 Dog Dumb
3 Cat Dumb
4 Cat Dumb
5 Dog Smart
6 Cat Smart """
```

```python
dframe = pd.read_table(StringIO(data),sep='\s+')
dframe
```

```python
pd.crosstab(dframe.Animal,dframe.Intelligence,margins=True)
```
