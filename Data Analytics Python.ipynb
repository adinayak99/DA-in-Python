```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
points = np.arange(-5,5,0.01)
dx,dy = np.meshgrid(points,points)

z = np.sin(dx) + np.sin(dy)
plt.title("Plot for sin(x) + sin(y)")
plt.imshow(z)
plt.colorbar()
```
